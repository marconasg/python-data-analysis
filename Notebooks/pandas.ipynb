__Analisar provas:__

```python
# Criar lista com o nome das provas
provas = ['math score', 'reading score', 'writing score']
```

```python
# Ordena o dataframe pela coluna provas
# Importante resetar o index para não ficar embaralhado
df = df.sort_values(by = provas, ascending = False)\
    .reset_index(drop = True)
```

```python
# Dataframe ordenado da maior nota pra menor nota
df
```

```python
# Adicionar uma coluna com a média das provas
df['mean'] = df[provas].mean(axis = 1)
```

```python
df.head()
```

```python
# Fazer consulta para visualizar os dados filtrados
# Quando a coluna tiver espaços entre as palavras, precisa escrever ela entre crases
df.query('(gender == "male") & (`test preparation course` == "none") & (`math score` >= 70)')
```

```python
# Outro método para filtrar os dados
df[(df.gender == 'male') & (df['test preparation course'] == 'none') & (df['math score'] >= 70)]
```

```python
# Mais um método para filtrar dados, agora usando o .loc
df.loc[(df.gender == 'male') & (df['test preparation course'] == 'none') & (df['math score'] >= 70)]
```

```python
# Agrupamento - agrupa os dados por gênero e obtém estatísticas descritivas
df.groupby(by = 'gender')[provas].agg([np.mean, np.median]).T
```
