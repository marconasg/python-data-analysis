## Seaborn
Seaborn é uma biblioteca de visualização de dados __Python__ baseada em __Matplotlib__. Ela fornece uma interface de alto nível para construir gráficos estatísticos atraentes e informativos.

__Importando a biblioteca:__

```python
import pandas as pd
import numpy as np
import seaborn as sns

# Definir tema
sns.set_theme(style = 'darkgrid')
```

__Base de dados de Restaurantes:__

```python
# Base de dados
baseDados = sns.load_dataset('tips')

# Mostrar base
baseDados.head()
```

```python
# Renomeado as colunas para PT-BR
baseDados.rename(columns = {
    'total_bill' : 'Total Gasto',
    'tip' : 'Gorjeta',
    'sex' : 'Sexo',
    'smoker' : 'Fumante',
    'day' : 'Dia da Semana',
    'time' : 'Período',
    'size' : 'Pessoas na Mesa'
}, inplace = True)
```

__Gráfico relplot:__

```python
sns.relplot( x = 'Total Gasto', y = 'Gorjeta', data = baseDados);
```

```python
# Gráfico relplot passando outro parâmetro como classe no gráfico
sns.relplot(x = 'Total Gasto', y = 'Gorjeta', data = baseDados, hue = 'Sexo');
```

```python
# Gráfico de linha com 2 eixos
sns.relplot(x = 'Total Gasto', y = 'Gorjeta', data = baseDados, kind = 'line');
```

```python
# Gráfico de linha com 2 eixos categorizando fumantes
sns.relplot(x = 'Total Gasto', y = 'Gorjeta', data = baseDados,
            kind = 'line', hue = 'Fumante');
```

__Gráfico histplot:__

```python
sns.histplot(data = baseDados, x = 'Total Gasto');
```

```python
sns.histplot(data = baseDados, x = 'Total Gasto', hue = 'Fumante');
```

__Gráfico barplot:__

```python
sns.barplot(data = baseDados, x = 'Sexo', y = 'Total Gasto', hue = 'Fumante');
```

__Gráfico pairplot:__

```python
sns.pairplot(baseDados);
```

```python
sns.pairplot(baseDados, hue = 'Sexo');
```

```python
sns.pairplot(baseDados, kind = 'kde');
```

__Gráfico boxplot:__

```python
sns.boxplot(data = baseDados, x = 'Dia da Semana', y = 'Total Gasto');
```

```python
sns.boxplot(data = baseDados, x = 'Dia da Semana', y = 'Total Gasto', hue = 'Sexo');
```

__Gráfico violinplot:__

```python
sns.violinplot(data = baseDados, x = 'Dia da Semana', y = 'Total Gasto');
```

__Gráfico swarmplot:__

```python
sns.swarmplot(data = baseDados, x = 'Total Gasto', y = 'Sexo');
```

__Gráfico jointplot:__

```python
sns.jointplot(data = baseDados, x = 'Total Gasto', y = 'Gorjeta');
```
